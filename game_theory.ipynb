## **پیادا سازی بازی معماس زندانی**

```python
import numpy as np
import itertools

# ماتریس پاداش برای هر بازیکن: [پاداش بازیکن 1، پاداش بازیکن 2]
# [همکاری، همکاری], [خیانت، همکاری], [همکاری، خیانت], [خیانت، خیانت]
payoff_matrix = {
    ('C', 'C'): (3, 3),  # هر دو همکاری کنند
    ('D', 'C'): (5, 0),  # بازیکن 1 خیانت کند، بازیکن 2 همکاری کند
    ('C', 'D'): (0, 5),  # بازیکن 1 همکاری کند، بازیکن 2 خیانت کند
    ('D', 'D'): (1, 1)   # هر دو خیانت کنند
}

# تابع برای محاسبه پاداش
def get_payoff(strategy1, strategy2):
    return payoff_matrix[(strategy1, strategy2)]

# استراتژی‌های ممکن برای بازیکنان
strategies = ['C', 'D']

# تعادل نش را پیدا می‌کنیم
def find_nash_equilibrium():
    for strategy1, strategy2 in itertools.product(strategies, repeat=2):
        # پاداش بازیکن 1 و بازیکن 2 برای استراتژی (strategy1, strategy2)
        payoff1 = get_payoff(strategy1, strategy2)[0]
        payoff2 = get_payoff(strategy1, strategy2)[1]

        # بررسی استراتژی بهینه برای بازیکن 1 و 2
        best_response1 = max([get_payoff(s, strategy2)[0] for s in strategies])
        best_response2 = max([get_payoff(strategy1, s)[1] for s in strategies])

        if payoff1 == best_response1 and payoff2 == best_response2:
            print(f"Nash Equilibrium: Player 1 -> {strategy1}, Player 2 -> {strategy2}")

# اجرای تابع
find_nash_equilibrium()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# تعریف ماتریس پاداش برای هر بازیکن: [پاداش بازیکن 1، پاداش بازیکن 2]
payoff_matrix = {
    ('C', 'C'): (3, 3),  # هر دو همکاری کنند
    ('D', 'C'): (5, 0),  # بازیکن 1 خیانت کند، بازیکن 2 همکاری کند
    ('C', 'D'): (0, 5),  # بازیکن 1 همکاری کند، بازیکن 2 خیانت کند
    ('D', 'D'): (1, 1)   # هر دو خیانت کنند
}

# تبدیل ماتریس پاداش به جدول Pandas
payoff_table = pd.DataFrame({
    'Player 1 / Player 2': ['C / C', 'D / C', 'C / D', 'D / D'],
    'Payoff (Player 1)': [payoff_matrix[('C', 'C')][0], payoff_matrix[('D', 'C')][0],
                          payoff_matrix[('C', 'D')][0], payoff_matrix[('D', 'D')][0]],
    'Payoff (Player 2)': [payoff_matrix[('C', 'C')][1], payoff_matrix[('D', 'C')][1],
                          payoff_matrix[('C', 'D')][1], payoff_matrix[('D', 'D')][1]]
})

# نمایش جدول
print("Payoff Table:")
print(payoff_table)

# رسم گراف چهارخانه‌ای و نمایش تعادل نش
def plot_payoff_matrix():
    fig, ax = plt.subplots()

    # ساختار مربع 2x2 برای نمایش چهار حالت مختلف
    payoff_values = np.array([
        [payoff_matrix[('C', 'C')], payoff_matrix[('C', 'D')]],
        [payoff_matrix[('D', 'C')], payoff_matrix[('D', 'D')]]
    ])

    # ایجاد جدول پاداش
    ax.matshow(np.zeros((2, 2)), cmap='Blues', alpha=0.3)

    # نمایش پاداش‌ها در هر خانه
    for i in range(2):
        for j in range(2):
            payoff1, payoff2 = payoff_values[i, j]
            ax.text(j, i, f"P1: {payoff1}\nP2: {payoff2}", va='center', ha='center', fontsize=12)

    # تنظیمات جدول
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_xticklabels(['C', 'D'])
    ax.set_yticklabels(['C', 'D'])
    ax.set_xlabel('Player 2 Strategy')
    ax.set_ylabel('Player 1 Strategy')

    # مشخص کردن تعادل نش
    nash_eq = (1, 1)  # موقعیت استراتژی (D, D) که تعادل نش است
    ax.plot(nash_eq[1], nash_eq[0], 'ro', markersize=15, label='Nash Equilibrium')

    # خطوط شبکه‌ای (grid) چهارخانه
    ax.grid(True, which='both', linestyle='-', linewidth=2)

    # نمایش لیبل‌ها و نمودار
    plt.legend()
    plt.title("Prisoner's Dilemma - Payoff Matrix and Nash Equilibrium")
    plt.show()

# نمایش جدول عایدی‌ها
payoff_table

# رسم نمودار با تعادل نش و خطوط شبکه‌ای
plot_payoff_matrix()
```
