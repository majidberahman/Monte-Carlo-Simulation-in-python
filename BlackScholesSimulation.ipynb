```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
T = 10
r = 1
sigma = 0.2
dt = 0.1
t = np.arange(0,T+dt,dt)
G = np.random.normal(0,1,len(t))
G[0] = 0
w = np.sqrt(dt)*np.cumsum(G)
plt.plot(t,w)
```
